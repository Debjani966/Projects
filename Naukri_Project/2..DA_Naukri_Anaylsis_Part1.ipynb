# Automation of the Company Analysis 

```python
#defining  afuction wil automize your code
# peparing a docstring for the function is easily readable and understandable by others

def perform_analysis(dataframe,feature,col1,col2,chart=1):
    
    '''
    dataframe: dataframe
    
    feature: column name that is consider for the analysis
    
    col1: name you want to assing to the column1
    
    col2: name you want to assing to the column2
    
    chart: chart=1 its represent bar chart
           chart=2 its represent pie chart
    
    '''
    
    data=dataframe[feature].value_counts().reset_index()
    data.columns=[col1,col2]
    plt.figure(figsize=(12,6))
    
    print(data.head(10))
    
    if chart==1:
        plt.bar(data[col1][0:10],data[col2][0:10])
        plt.xticks(rotation='vertical')
    elif chart==2:
        plt.pie(labels=data[col1][0:10],x=data[col2][0:10])
        plt.xticks(rotation='vertical')
```

```python
perform_analysis(data,'company','company_name','Number of Jobs',chart=2) #shift+tab = will give you the docstr
```

```python
perform_analysis(data,'company','company_name','Number of Jobs',chart=1) #shift+tab = will give you the docstring
```
